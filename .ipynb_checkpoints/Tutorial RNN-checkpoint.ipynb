### 1.3.1 A feed-forward example

```python
w1 = np.array([[0.2,0.2,0.2], [0.4,0.4,0.4],[0.6,0.6,0.6]])
w1
```

```python
b1 = np.array([0.8, 0.8, 0.8])
b2 = np.array([0.2])
```

```python
def f(x):
    return 1/(1+np.exp(-x))
```

```python
def simple_looped(n_layers, x,w,b):
    for l in range(n_layers-1):
        if l == 0:
            node_in = x
        else:
            node_in = h
        h = np.zeros((w[l].shape[0],))  
        
        for i in range(w[l].shape[0]):
            f_sum = 0
            for j in range(w[l].shape[1]):
                f_sum += w[l][i][j] * node_in[j]
            f_sum += b[l][i]                
            h[i] = f(f_sum)                
                                 
    return h
```

```python
w = [w1,w2]
b = [b1,b2]
x = [1.5, 2.0, 3.0]
```

```python
simple_looped(3,x,w,b)
```

```python
%time simple_looped(3,x,w,b)
```

### Usando numpy 

```python
def matrix_feedforward(n_layer, x,w,b):
    for l in range(n_layer-1):
        if l == 0:
            node_in = x
        else:
            node_in = h
        z = w[l].dot(node_in) + b[l]   
        h = f(z)
    return h
        
```

```python
%time matrix_feedforward(3,x,w,b)
```

### Gradien descent

```python
x_old = 0
x_new = 6
gamma = 0.01
precision = 0.00001

def df(x):
    y = 4 * x**3 - 9 * x**2
    return y

while abs(x_new-x_old) > precision:
    x_old = x_new
    x_new += -gamma * df(x_old)
    
print("the local minimum occurs at %f"% x_new)
```

### Creando la red neuronal

```python
nn_structure = [64,30,10]
len(nn_structure)
```

```python
def f(x):
    return 1/(1 + np.exp(-x))
```

```python
def f_derive(x):
    return f(x)*(1-f(x))
```

```python
def inicialize_w(nn_structure):
    w = {}
    b = {}
    for l in range(1, len(nn_structure)):
        w[l] = np.random.random_sample((nn_structure[l], nn_structure[l-1]))
        b[l] = np.random.random_sample((nn_structure[l],))
    return w,b
```

```python
nn_structure[2]
```

```python
np.random.random_sample((30, 10))
```

```python
def init_tri_val(nn_structure):
    tri_w = {}
    tri_b = {}
    for l in range(1, len(nn_structure)):
        tri_w[l] = np.zeros((nn_structure[l], nn_structure[l-1]))
        tri_b[l] = np.zeros((nn_structure[l],))
    return tri_w, tri_b
```

```python
def feed_forward(x,w,b):
    h = {1:x}
    z = {}
    for l in range(1, len(w)+1):
        if l == 1:
            node_in = x
        else:
            node_in = h[l]
        z[l+1] = w[l].dot(node_in) + b[l]
        h[l+1] = f(z[l+1])
    return h, z
        
        
```

### back propagation 

```python
def calculate_outlayer_delta(y, h_out, z_out):
    return -(y-h_out) * f_derive(z_out)
```

```python
def calculate_hidden_delta(delta_plus1, w_1, z_1):
    return np.dot(np.transpose(w_1), delta_plus1) * f_derive(z_1)
```

```python
def train_nn(nn_structure, X,y, iter_num=3000, alpha=0.25):
    w,b = inicialize_w(nn_structure)
    cnt = 0
    m = len(y)
    avg_cost_function = []
    print ('starting gradient descent for{} iterations'.format(iter_num))
    while cnt < iter_num:
        if cnt%1000 == 0:
            print ('iteration {} of {}'.format(cnt,iter_num))
        tri_w, tri_b = init_tri_val(nn_structure)
        avg_cost = 0
        for i in range(len(y)):
            delta = {}
            
            h,z = feed_forward(X[i,:],w,b)
            for l in range(len(nn_structure),0,-1):
                if l == len(nn_structure):
                    delta[l] = calculate_outlayer_delta(y[i,:], h[l], z[l])
                    avg_cost += np.linalg.norm((y[i,:]-h[l]))
                else:
                    if l > 1:
                        delta[l] = calculate_hidden_delta(delta[l+1], w[l], z[l])
                    tri_w[l] += np.dot(delta[l+1][:,np.newaxis], np.transpose(h[l][:,np.newaxis])) 
                    tri_b[l] += delta[l+1]
        
        for l in range(len(nn_structure)-1,0,-1):
            w[l] += -alpha * (1.0/m * tri_w[l])
            b[l] += -alpha * (1.0/m * tri_b[l])
            
        avg_cost = 1.0/m * avg_cost
        avg_cost_function.append(avg_cost)
        cnt += 1
        
    return w,b, avg_cost_function
            
```

```python
w,b,avg_cost_function = train_nn(nn_structure, x_train, y_vect_train)
```

```python
fig = plt.figure(figsize=(12,8))
plt.plot(avg_cost_function)
plt.ylabel('Average J')
plt.xlabel('iteration number')
plt.show()
```

```python
def predict_y(w,b,X,n_layers):
    m = X.shape[0]
    y = np.zeros((m,))
    for i in range(m):
        h,z = feed_forward(X[i,:],w,b)
        y[i] = np.argmax(h[n_layers])
    return y
```

```python
from sklearn.metrics import accuracy_score
y_pred = predict_y(w,b,x_test,3)
accuracy_score(y_test, y_pred)*100
```
